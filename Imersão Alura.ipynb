````python
df_ticker = pd.read_excel("/content/drive/MyDrive/IT/Dados/Imersão Alura - Python/Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
display(df_ticker)
````

````python
df_total_acoes = pd.read_excel("/content/drive/MyDrive/IT/Dados/Imersão Alura - Python/Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
display(df_total_acoes)
````

````python
df_chat_gpt = pd.read_excel("/content/drive/MyDrive/IT/Dados/Imersão Alura - Python/Imersão Python - Tabela de ações.xlsx", sheet_name="ChatGPT")
display(df_chat_gpt)
````

Preparar base de dados e visualizar resumo de dados.

````python
df_principal = df_principal[['Ativo', 'Último (R$)', 'Var. Dia (%)']].copy()
display(df_principal)
````

````python
display(df_principal.info())
````

Renomear colunas.

````python
df_principal = df_principal.rename(columns={'Último (R$)':'Valor_final','Var. Dia (%)':'Var_dia_pct'}).copy()
display(df_principal)
````

# Merge

````python
df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')
df_principal = df_principal.rename(columns={'Qtde. Teórica': 'Qtd_teorica'})
display(df_principal)
````

````python
df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')
display(df_principal)
````

````python
#Formatar valores tipo float
pd.options.display.float_format = '{:.2f}'.format

#Formatar valores tipo int

df_principal['Qtd_teorica'] = df_principal['Qtd_teorica'].astype(int)
````

````python
df_principal = df_principal.merge(df_chat_gpt, left_on='Nome', right_on='Nome da Empresa', how='left')
df_principal = df_principal.rename(columns={'Nome da Empresa': 'Nome_empresa'})
display(df_principal)
````

Remover colunas duplicadas.

````python
df_principal = df_principal.drop(columns=['Código', 'Ticker', 'Nome'])
display(df_principal)
````

# Cálculos

````python
df_principal['Var_pct'] = df_principal['Var_dia_pct'] / 100
display(df_principal)
````

````python
df_principal['Valor_inicial'] = df_principal['Valor_final'] / (df_principal['Var_pct'] + 1 )
display(df_principal)
````

````python
df_principal['Variacao_rs'] = (df_principal['Valor_final'] - df_principal['Valor_inicial']) * df_principal['Qtd_teorica']
display(df_principal)
````

````python
df_principal['Resultado'] = df_principal['Variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estável'))
df_principal = df_principal.rename(columns={'Idade (anos)':'Idade_anos'})
display(df_principal)
````

````python
def faixa_idade(x):
     if x <= 40:
         return '20 a 40 anos'
     elif x <= 60:
         return '41 a 60 anos'
     else:
         return 'Mais de 60 anos'

df_principal['Faixa_idade'] = df_principal['Idade_anos'].apply(faixa_idade)
display(df_principal)
````

# Análises

````python
#Calculando o maior valor
maior = df_principal['Variacao_rs'].max()

#Calculando o menor valor
menor = df_principal['Variacao_rs'].min()

#Calculando a media total
media = df_principal['Variacao_rs'].mean()

#Calculando a media de quem subiu
media_subiu = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_rs'].mean()

#Calculando a media de quem desceu
media_desceu = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_rs'].mean()

#Imprimindo resultados

print(f"Maior\tR$ {maior:,.2f}")
print(f"Menor\tR$ {menor:,.2f}")
print(f"Media\tR$ {media:,.2f}")
print(f"Média de quem subiu\tR$ {media_subiu:,.2f}")
print(f"Média de quem desceu\tR$ {media_desceu:,.2f}")
````

````python
df_principal_subiu = df_principal[df_principal['Resultado'] == 'Subiu']
display(df_principal_subiu)
````

````python
df_analise_segmento = df_principal_subiu.groupby('Segmento')['Variacao_rs'].sum().reset_index()
display(df_analise_segmento)
````

````python
df_analise_saldo = df_principal.groupby('Resultado')['Variacao_rs'].sum().reset_index()
display(df_analise_saldo)
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import mplfinance as mpf
import yfinance as yf
import plotly.graph_objects as go
from plotly.subplots import make_subplots
````

# Base de dados

````python
dados = yf.download('PETR4.SA', start='2023-01-01', end='2023-12-31')
display(dados)
````

````python
dados.columns = ['Abertura', 'Maximo', 'Minimo', 'Fechamento', 'Fech_ajus', 'Volume'] # Nomear as colunas
dados = dados.rename_axis('Data')
display(dados)
````

# Gráficos

````python
dados['Fechamento'].plot(figsize=(15,6))
plt.legend('Fechamento')
plt.title('Variação do Preço por Data', fontsize=16)
````

````python
df =dados.head(60).copy()

df['Data'] = df.index

df['Data'] = df['Data'].apply(mdates.date2num)
display(df)
````

````python
fig, ax = plt.subplots(figsize=(15, 8))


width = 0.7  # Definir largura


for i in range(len(df)):
    if df['Fechamento'].iloc[i] > df['Abertura'].iloc[i]:
        color = 'green'
    else:
        color = 'red'


    ax.plot([df['Data'].iloc[i], df['Data'].iloc[i]],
            [df['Minimo'].iloc[i], df['Maximo'].iloc[i]],
            color=color,
            linewidth=1)


    ax.add_patch(plt.Rectangle((df['Data'].iloc[i] - width / 2, min(df['Abertura'].iloc[i], df['Fechamento'].iloc[i])),
                                width,
                                abs(df['Fechamento'].iloc[i] - df['Abertura'].iloc[i]),
                                facecolor=color))


df['MA7'] = df['Fechamento'].rolling(window=7).mean()
df['MA14'] = df['Fechamento'].rolling(window=14).mean()


ax.plot(df['Data'], df['MA7'], color='orange', label='Média Móvel 7 Dias')  # Média de 7 dias
ax.plot(df['Data'], df['MA14'], color='yellow', label='Média Móvel 14 Dias')  # Média de 14 dias


ax.legend()


# Formatando o eixo x para mostrar as datas
# Configuramos o formato da data e a rotação para melhor legibilidade
ax.xaxis_date()  # O método xaxis_date() é usado para dizer ao Matplotlib que as datas estão sendo usadas no eixo x
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)


plt.title("Gráfico de Candlestick - PETR4.SA com matplotlib")
plt.xlabel("Data")
plt.ylabel("Preço")


plt.grid(True)


plt.show()

````

# Prophet

O Prophet é uma ferramenta desenvolvida pelo Facebook para fazer previsões em séries temporais, que são basicamente dados coletados ao longo do tempo, como os preços das ações. É famoso por ser fácil de usar e por funcionar bem mesmo com poucos dados ou com dados que possuem padrões irregulares, o que é comum no mercado de ações.

Vamos usar dados de ações da empresa Johnson & Johnson como exemplo.

Aqui, usamos um comando para baixar os dados.

````python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
from prophet import Prophet
````

````python
# Baixar dados
dados = yf.download("JNJ", start="2020-01-01", end="2023-12-31", progress=False)
dados = dados.reset_index()
````

````python
# Dados em treino (até o final do primeiro semestre de 2023) e Dados em teste (segundo semestre de 2023)
dados_treino = dados[dados['Date'] < '2023-07-31']
dados_teste = dados[dados['Date'] >= '2023-07-31']
````

````python
# Preparando os dados
dados_prophet_treino = dados_treino[['Date', 'Close']].rename(columns={'Date': 'ds', 'Close': 'y'})
````

````python
# Criar e treinar o modelo
modelo = Prophet(weekly_seasonality=True,
        yearly_seasonality=True,
        daily_seasonality=False)

modelo.add_country_holidays(country_name='BR')

modelo.fit(dados_prophet_treino)
````

````python
# Criar datas futuras para previsão até o final de 2023
futuro = modelo.make_future_dataframe(periods=150)
previsao = modelo.predict(futuro)
````

````python
plt.figure(figsize=(14, 8))
plt.plot(dados_treino['Date'], dados_treino['Close'], label='Dados de Treino', color='blue')
plt.plot(dados_teste['Date'], dados_teste['Close'], label='Dados Reais (Teste)', color='green')
plt.plot(previsao['ds'], previsao['yhat'], label='Previsão', color='orange', linestyle='--')

plt.axvline(dados_treino['Date'].max(), color='red', linestyle='--', label='Início da Previsão')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento')
plt.title('Previsão de Preço de Fechamento vs Dados Reais')
plt.legend()
plt.show()
````
